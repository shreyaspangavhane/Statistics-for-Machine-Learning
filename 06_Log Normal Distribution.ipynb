```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df=pd.read_csv("income1.csv",index_col=None,names=["income","count"],skiprows=1)
df.head()
```

```python
plt.figure.figsize=(15,8)
g=sns.barplot(x="income",y="count",data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45)
```

```python
plt.figure.figsize=(15,8)
g=sns.barplot(x="income",y="count",data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45)
g.set(xscale="log")
```
