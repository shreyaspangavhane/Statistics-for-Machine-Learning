```python
df['revenue_mln']=df["revenue"].apply(lambda x:x/1000000)     # add col revenue in million and assign value from revenue value/10^6
df.revenue_mln.describe()
```

```python
_,mean,std,*_ =df.revenue_mln.describe()                        #make the tuple and use underscore for that which i don't need to that value
```

```python
mean,std
```

```python
# make the functuon which will calculate the value of the Z score for all records
def get_Z_score(value,mean,std):
    return (value-mean)/std
```

```python
df["z_score"]=df.revenue_mln.apply(lambda x:get_Z_score(x,mean,std))
```

```python
df.head()
```

```python
# check outlier =>> z_score>3 is the outlier
df.z_score>3
```

```python
# to get only that record
df[df.z_score>3]
```

```python
## to remove that outlier
df_no_outlier=df[df.z_score<3]
df_no_outlier.head()
```

## Modified Zscore to detect outlier

```python
## calculate Mean Absolute deviation       formula=>> median(s-median(s))
def get_MAD(s):
    median=np.median(s)
    diff=abs(s-median)
    MAD=np.median(diff)
    return MAD
```

```python
MAD=get_MAD(df.revenue_mln)             ## cal MAD of revenue of million
median=np.median([df.revenue_mln.median()])
MAD,median
```

```python
## Calculate the modified Z score
def get_modified_z_score(x,median,MAD):
    return 0.6745*((x-median)/MAD)
```

```python
## Calculate the modified Z score for AVATAR Movie
get_modified_z_score(2787.96,median,MAD)              # 2787.96 is the  revenue in million
```

```python
## Calculate the modified Z score for ALL Movie

df["modified_z_score"]=df.revenue_mln.apply(lambda x:get_modified_z_score(x,median,MAD))
df.head()
```

```python
df[df["modified_z_score"]>3.5]        # get the 3 outlier before we get only one Outlier when we use the Z score 
```

# Observation:
<h3>When we use the Z score we get only one outlier but when we  use the Modified Z Score we get 3 Outlier</h3>
